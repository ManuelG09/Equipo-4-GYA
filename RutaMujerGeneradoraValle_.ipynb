```python

# Mostrar todas las hojas para verificar
print("Hojas disponibles:")
for i, hoja in enumerate(hojas):
    print(f"{i+1}: {hoja}")

# Seleccionar la hoja número 7 (índice 6)
indice_hoja = 6

# Verificar si existe esa hoja
if indice_hoja < len(hojas):
    nombre_hoja = hojas[indice_hoja]
    print(f"Hoja seleccionada: {nombre_hoja}")

    # Leer la hoja 7
    df_hoja7 = pd.read_excel(ruta_archivo, sheet_name=nombre_hoja)


    # Mostrar las primeras filas
    df_hoja7.head()
else:
    print(f"No hay una hoja número 7. Solo hay {len(hojas)} hojas.")
```

```python
# Leer directamente la hoja número 7 (índice 6)
df_Valle = pd.read_excel(ruta_archivo, sheet_name=6)

# Leer la hoja 7 (índice 6) y saltar las primeras 3 filas
df_Valle = pd.read_excel(ruta_archivo, sheet_name=6, skiprows=4)

# Mostrar las primeras filas
df_Valle.head()
```

```python
df_Valle.info()
```

```python
# Quitar columnas donde todos los valores son NaN
df_Valle_limpio = df_Valle.dropna(axis=1, how='all')

# Mostrar columnas resultantes
print(df_Valle_limpio.columns)
```

```python
# Limpiar nombres de columnas (quitar saltos de línea y espacios extra)
df_Valle_limpio.columns = df_Valle_limpio.columns.str.replace('\n', ' ', regex=False).str.strip()
df_Valle_limpio.columns = df_Valle_limpio.columns.str.replace('  ', ' ', regex=False)  # Quitar dobles espacios

# Verificar resultado
print(df_Valle_limpio.columns.tolist())
```

```python
df_Valle_limpio = df_Valle_limpio.rename(columns={
    'RIPSFinalidadConsultaVersion2 RESOLUCION 2275-2023': 'RIPS Resolucion 2275_2023',
    'NIVEL DE COMPLEJIDA B- M -A': 'NIVEL_COMPLEJIDAD',
    'MODALIDAD E-C': 'MODALIDAD'
})
```

```python
# Renombrar la columna 'FINALIDAD_CODIGO' por 'RIPS Resolucion 2275_2023''
df_Valle_limpio = df_Valle_limpio.rename(columns={
    'FINALIDAD_CODIGO': 'RIPS Resolucion 2275_2023'
})

# Verificar nombres de columnas
print(df_Valle_limpio.columns.tolist())
```

```python
# Convertir la columna FINALIDAD a tipo numérico
# Los errores (valores no convertibles) se ponen como NaN
df_Valle_limpio['FINALIDAD'] = pd.to_numeric(df_Valle_limpio['FINALIDAD'], errors='coerce').fillna(0).astype(int)


# Reemplazar NaN por 0
df_Valle_limpio['FINALIDAD'] = df_Valle_limpio['FINALIDAD'].fillna(0)

# Verificar tipo de dato y valores
print(df_Valle_limpio['FINALIDAD'].dtype)
print(df_Valle_limpio['FINALIDAD'].head())
```

```python
# Lista de columnas a modificar
columnas_a_reemplazar = [
    'DESCRIPCION CUPS',
    'DX RELACIONADO',
    'CRITERIO PARA LA PRESTACION DEL SERVICIO'
]

# Reemplazar NaN por 'NINGUNO' en esas columnas
df_Valle_limpio[columnas_a_reemplazar] = df_Valle_limpio[columnas_a_reemplazar].fillna('NINGUNO')

# Verificar que se hizo bien
print(df_Valle_limpio[columnas_a_reemplazar].head())
```
