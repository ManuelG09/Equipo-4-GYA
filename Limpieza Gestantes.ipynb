```python
#Función para limpiar texto: eliminar espacios extra, quitar tildes y pasar a mayúsculas
def limpiar_texto(texto):
    if isinstance(texto, str):  # Verifica si es texto
        texto = texto.strip()  # Elimina espacios al inicio y al final
        texto = re.sub(r'\s+', ' ', texto)  # Reemplaza múltiples espacios por uno solo
        texto = unidecode(texto)  # Quita tildes
        return texto.upper()  # Convierte a mayúsculas
    return texto  # Si no es texto, lo deja igual
df_data2= df_data.applymap(limpiar_texto)
```

```python
# Función para estandarizar fechas
def estandarizar_fecha(fecha):
    try:
        # Intentar con formato ISO (YYYY-MM-DD HH:MM:SS.sss)
        return pd.to_datetime(fecha, format="%Y-%m-%d %H:%M:%S.%f").strftime("%d/%m/%Y")
    except ValueError:
        try:
            # Intentar con formato YYYY-MM-DD HH:MM:SS
            return pd.to_datetime(fecha, format="%Y-%m-%d %H:%M:%S").strftime("%d/%m/%Y")
        except ValueError:
            try:
                # Intentar con formato DD/MM/YYYY
                return pd.to_datetime(fecha, format="%d/%m/%Y").strftime("%d/%m/%Y")
            except ValueError:
                return None  # Si no coincide con ningún formato, devolver None

# Aplicar la conversión
df_data3["FECHA_IDENTIFICACION"] = df_data3["FECHA_IDENTIFICACION"].apply(estandarizar_fecha)
# Convertir a tipo fecha en Pandas
df_data3["FECHA_IDENTIFICACION"] = pd.to_datetime(df_data3["FECHA_IDENTIFICACION"], format="%d/%m/%Y")
```

```python
# Función para limpiar y normalizar Peso
def limpiar_peso(peso):
    peso = str(peso)
    PESOINGRESO = peso.strip()
    PESOINGRESO = PESOINGRESO.replace(',', '.')
    PESOINGRESO = ''.join(filter(lambda x: x.isdigit() or x == '.', PESOINGRESO))
    try:
        return float(PESOINGRESO)
    except ValueError:
        return np.nan
df_data3['PESOINGRESO'] = df_data3['PESOINGRESO'].apply(limpiar_peso)

print(df_data3['PESOINGRESO'])

```

```python
# Función para limpiar y normalizar Talla
def limpiar_talla(talla):
    talla = str(talla)
    TALLA = talla.strip()
    TALLA = TALLA.replace(',', '.')
    TALLA = ''.join(filter(lambda x: x.isdigit() or x == '.', TALLA))
    try:
        return float(TALLA)
    except ValueError:
        return np.nan

df_data3['TALLA'] = df_data3['TALLA'].apply(limpiar_talla)

print(df_data3['TALLA'])
```
