```python
#Corregir departamento por el codigo DIVIPOLA
df_data2["CODIGO_DEP_PRE_SER"] = df_data2["CODIGO_DEP_PRE_SER"].str.replace(r"\.0$", "", regex=True)  # Quitar ".0"
df_data2["CODIGO_DEP_PRE_SER"] = df_data2["CODIGO_DEP_PRE_SER"].apply(lambda x: "0" + x if len(x) == 1 else x)  # Agregar '0' si tiene 1 dígitos
```

```python
#Corregir IPS_ASIGNADA para que el codigo de habilitacion quede okey
df_data2["IPS_ASIGNADA"] = df_data2["IPS_ASIGNADA"].str.replace(r"\.0$", "", regex=True)  # Quitar ".0"
df_data2["IPS_ASIGNADA"] = df_data2["IPS_ASIGNADA"].apply(lambda x: "0" + x if len(x) == 11 else x)  # Agregar '0' si tiene 11 dígitos

```
